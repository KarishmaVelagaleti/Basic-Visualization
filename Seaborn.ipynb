## DATA VISUALIZATION : SEABORN LIBRARY

```python
import seaborn as sns
%matplotlib inline # to load charts on your notebook
import matplotlib.pyplot as plt
```

```python
# SAMPLE DATA SET
tips = sns.load_dataset('tips')
tips.head()
```

### Histogram

```python
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'], kde=False, color='red')
```

```python
sns.distplot(tips['total_bill'], kde=False, color='red', bins=20)
```

### JOINTPLOT

```python
sns.jointplot(x='total_bill', y='tip', data=tips)
```

```python
sns.jointplot(x='total_bill', y='tip',kind='hex', data=tips)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde')
```

### RUGPLOT

```python
sns.rugplot(tips["total_bill"])
```

### CATEGORICAL PLOTS

##### BARPLOT

```python
sns.barplot(x='sex', y='total_bill', data=tips) ## shows average values
```

```python
import numpy as np
```

```python
sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.mean)
```

##### COUNTPLOT

```python
sns.countplot(x='sex', data=tips)
```

##### BOX PLOTS

```python
sns.boxplot(x='day', y='total_bill', data=tips)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, hue='smoker')
```

##### VIOLIN PLOTS 
(down-side = hard to interpret and read. Not recommended for executive presentation)


```python
sns.violinplot(x='day', y='total_bill', data=tips)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex',split=True)
```

##### STRIP PLOT

```python
sns.stripplot(x='day', y="total_bill", data=tips)
```

```python
sns.stripplot(x='day', y="total_bill", data=tips, jitter=True)
```

##### SWARMPLOT

Again - hard to interpret

```python
sns.swarmplot(x='day',y="total_bill", data=tips)
```

##### FACTORPLOT

```python
sns.factorplot(x='day', y='total_bill', data=tips, kind='bar')
```

```python
sns.factorplot(x='day', y='total_bill', data=tips, kind='violin')
```

## NUMERICAL PLOTS

### Heat MAP

```python
sns.heatmap(tips.corr())
```

```python
sns.heatmap(tips.corr(), annot=True)
```

#### PIVOT TABLE

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
flights.pivot_table(index='month', columns='year', values='passengers')
```

```python
fp=flights.pivot_table(index='month', columns='year', values='passengers')
```

```python
plt.figure(figsize=(10,7))
sns.heatmap(fp, cmap='magma', linecolor='white', linewidth=1)
```

### CLUSTERMAP

very useful when you want to use 'clustering'

```python
plt.figure(figsize=(12,10))
sns.clustermap(fp, cmap='magma')
```

### Pair plot

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips, hue='sex')
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.pairplot(iris)
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

## FactGrid Plot

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
g = sns.FacetGrid(data=tips, col='time', row='smoker')
g.map(sns.distplot, 'total_bill')
```

```python
g = sns.FacetGrid(data=tips, col='time', row='smoker')
g.map(plt.scatter, 'total_bill', 'tip')
```

## Linear Model Plots or Regression Plots

```python
sns.lmplot(x='total_bill', y='tip', data=tips)
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['*', 'v'])
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', col='sex')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', row='sex')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', col='sex', row='time')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', col='sex', row='time', aspect=0.6,size=8)
```
